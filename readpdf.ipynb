```python

# filename: readpdy,ipynb
# created: 04/2023
# porpuse: part of the capstone project for Data Science track
# reads pdf files downloaded from www.cirr.org/data to extract variables to be analized
# generates a cvs files with variables to be analized

# importing required modules
import os
from PyPDF2 import PdfReader
import pandas as pd

#assign constant values to variables to look for whitin the file
v1 = 'School Name'
v2 = 'Campus Location'
v3 = 'Program Name'
v4 = 'Reporting Period'
v5 = 'Published Course Length (in days, including weekends and holidays)'
v6 = 'Graduates Included in Report'
v7 = 'Graduation Requirements'
v8 = 'How many students graduate within 100% of published program length (on-time)' #?
v9 = 'How many students graduate within 150% of published program length' #?
v10 = 'Job Seekers'
v11 = 'How many students intended to seek in-field employment within 180 days of graduating?'
v12 = 'How many students did not intend to seek in-field employment (returning to previous'
v13 = 'employer, no work authorization, continuing to further education, or self-enrichment)?'
v14 = 'Employment Results'
v15 = '1. Employed in-field'
v16 = '1A. Full-time employee (30+ hours/week, 6+ months)'
v17 = '1B. Full-time apprenticeship, internship, or contract position (30+ hours/week, 3-6 months)'
v18 = '1C. Short-term contract, part-time position, freelance, or unknown length'
v19 = '1D. Started a new company or venture after graduation'
v20 = '2. Not seeking in-field employment'
v21 = '2A. Employed out-of-field'
v22 = '2B. Continuing to higher education'
v23 = '2C. Not seeking a job for health, family, or personal reasons'
v24 = '3. Still seeking a job in-field'
v25 = '4. Could not contact'
v26 = 'Hired by School'
v27 = 'Median Annual Base Salary'
v27b = 'What is the median annual base salary of graduates?'
vrange50_u50 = 'Under $50,000'
vrange60_u60 = 'Under $60,000'
vrange70_u70 = 'Under $70,000'
v40_50 = '$40,000-$50,000'
v50_60 = '$50,000-$60,000'
v60_70 = '$60,000-$70,000'
v70_80 = '$70,000-$80,000'
v80_90 = '$80,000-$90,000'
v90_100 = '$90,000-$100,000'
v100_110 = '$100,000-$110,000'
vrange_40_o80 = 'Over $80,000'
vrange_50_o90 = 'Over $90,000'
vrange_60_o100 = 'Over $100,000'
vrange_70_o110 = 'Over $110,000'
v34 = 'Percentage of job obtainers who reported salaries'
v35 = 'Most Frequent Job Titles'

def splitVariable(v, printflag):
    if printflag:
        print(v)
    l = v.split()
    if len(l) == 2:
        return l[0], l[1]
    else:
        return "1", "2"

def sumAndPercentage(v):
    sum0 = 0.0
    for e in v:
        sum0 += float(e)
    return str(sum0) + '%'


def replaceAndSum(v):
    vl = v.split()
    vl[0] = vl[0].replace('%', ' ')
    vl[1] = vl[1].replace('%', ' ')

    vl0 = sumAndPercentage(vl[0].split())
    vl1 = sumAndPercentage(vl[1].split())

    return vl0 + ' ' + vl1


def parse_text(filename, text):
    # look for variables within the file (the text)
    pos1 = text.find(v1)
    pos2 = text.find(v2)
    pos3 = text.find(v3)
    pos4 = text.find(v4)
    pos5 = text.find(v5)
    pos6 = text.find(v6)
    pos7 = text.find(v7)
    pos8 = text.find(v8)
    pos9 = text.find(v9)
    pos10 = text.find(v10)
    pos11 = text.find(v11)
    pos12 = text.find(v12)
    pos13 = text.find(v13)
    pos14 = text.find(v14)
    pos15 = text.find(v15)
    pos16 = text.find(v16)
    pos17 = text.find(v17)
    pos18 = text.find(v18)
    pos19 = text.find(v19)
    pos20 = text.find(v20)
    pos21 = text.find(v21)
    pos22 = text.find(v22)
    pos23 = text.find(v23)
    pos24 = text.find(v24)
    pos25 = text.find(v25)
    pos26 = text.find(v26)
    pos27 = text.find(v27)

    flagv27b = 0

    if pos27 == -1:
        print("Didn't find v27", v1)
        pos27 = text.find(v27b)
        flagv27b = 1

    if text.find(vrange50_u50) > 0:
        v28 = vrange50_u50
        v29 = v50_60
        v30 = v60_70
        v31 = v70_80
        v32 = v80_90
        v33 = vrange_50_o90

    elif text.find(vrange60_u60) > 0:
        v28 = vrange60_u60
        v29 = v60_70
        v30 = v70_80
        v31 = v80_90
        v32 = v90_100
        v33 = vrange_60_o100

    elif text.find(vrange70_u70) > 0:
        v28 = vrange70_u70
        v29 = v70_80
        v30 = v80_90
        v31 = v90_100
        v32 = v100_110
        v33 = vrange_70_o110

    pos28 = text.find(v28)
    pos29 = text.find(v29)
    pos30 = text.find(v30)
    pos31 = text.find(v31)
    pos32 = text.find(v32)
    pos33 = text.find(v33)

    pos34 = text.find(v34)
    pos35 = text.find(v35)

    schoolName     = text[pos1+len(v1)+1:pos2-1]
    campusLocation = text[pos2+len(v2)+1:pos3-1]
    programName    = text[pos3+len(v3)+1:pos4-1]

    reportingPeriod = text[pos4+len(v4)+1:pos5-1]
    reportingPeriodStart, reportingPeriodEnd = splitVariable(reportingPeriod, 0)

    publishedCourseLengthWeeks = text[pos5+len(v5)+1:pos6-1]
    graduatesIncluded = text[pos6+len(v6)+1:pos7-1]
    studentsGraduateWithinProgramLength = text[pos8+len(v8)+1:pos9-1]
    studentsIntendedInFieldEmployment = text[pos11+len(v11)+1:pos12-1]

    employedInField = text[pos15+len(v15)+1:pos16-1]
    employedInField90d, employedInField180d = splitVariable(employedInField, 0)
    
    notSeekingInField = text[pos20+len(v20)+1:pos21-1]
    notSeekingInField90d, notSeekingInField180d = splitVariable(notSeekingInField, 0)

    stillSeekingInField = text[pos24+len(v24)+1:pos25-1]
    stillSeekingInField90d, stillSeekingInField180d = splitVariable(stillSeekingInField, 0)

    if flagv27b == 1:
        medianAnnualBaseSalary = text[pos27+len(v27b)+1:pos28-1]
    else:
        medianAnnualBaseSalary = text[pos27+len(v27)+1:pos28-1]

    medianAnnualBaseSalary90d, medianAnnualBaseSalary180d = splitVariable(medianAnnualBaseSalary, 0)

    if text.find(vrange50_u50) > 0:
        u50 = text[pos28+len(v28)+1:pos29-1]
        _50_60 = text[pos29+len(v29)+1:pos30-1]
        _60_70 = text[pos30+len(v30)+1:pos31-1]

        u5090, u50180 = splitVariable(u50, 0)
        _506090, _5060180 = splitVariable(_50_60, 0)
        _607090, _6070180 = splitVariable(_60_70, 0)

        u70 = u5090 + _506090 + _607090 + ' ' + u50180 + _5060180 + _6070180
        u70 = replaceAndSum(u70)

        _70_80 = text[pos31+len(v31)+1:pos32-1]
        _80_90 = text[pos32+len(v32)+1:pos33-1]

        o90 = text[pos33+len(v33)+1:pos34-1]

    elif text.find(vrange60_u60) > 0:
        u60 = text[pos28+len(v28)+1:pos29-1]
        _60_70 = text[pos29+len(v29)+1:pos30-1]

        u6090, u60180 = splitVariable(u60, 0)
        _607090, _6070180 = splitVariable(_60_70, 0)

        u70 = u6090 + _607090 + ' ' + u60180 + _6070180
        u70 = replaceAndSum(u70)

        _70_80 = text[pos30+len(v30)+1:pos31-1]
        _80_90 = text[pos31+len(v31)+1:pos32-1]

        _90_100 = text[pos32+len(v32)+1:pos33-1]
        o100 = text[pos33+len(v33)+1:pos34-1]

        _9010090, _90100180 = splitVariable(_90_100, 0)
        o10090, o100180 = splitVariable(o100, 0)

        o90 = _9010090 + o10090 + ' ' + _90100180 + o100180
        o90 = replaceAndSum(o90)

    elif text.find(vrange70_u70) > 0:
        u70 = text[pos28+len(v28)+1:pos29-1]

        _70_80 = text[pos29+len(v29)+1:pos30-1]
        _80_90 = text[pos31+len(v31)+1:pos32-1]

        _90_100 = text[pos31+len(v31)+1:pos32-1]
        _100_110 = text[pos32+len(v32)+1:pos33-1]
        o110 = text[pos33+len(v33)+1:pos34-1]

        _9010090, _90100180 = splitVariable(_90_100, 0)
        _10011090, _100110180 = splitVariable(_100_110, 0)
        o11090, o110180 = splitVariable(o110, 0)

        o90 = _9010090 + _10011090 + o11090 + ' ' + _90100180 + _100110180 + o110180
        o90 = replaceAndSum(o90)

    under70K = u70
    under70K90d, under70K180d = splitVariable(under70K, 0)

    _70K_80K = _70_80
    _70K_80K90d, _70K_80K180d = splitVariable(_70K_80K, 0)

    _80K_90K = _80_90
    _80K_90K90d, _80K_90K180d = splitVariable(_80K_90K, 0)

    over90K = o90
    over90K90d, over90K180d = splitVariable(over90K, 0)

    myrow = [
#    filename,
    schoolName,
    campusLocation,
    programName,
    reportingPeriodStart,
    reportingPeriodEnd,
    publishedCourseLengthWeeks,
    graduatesIncluded,
    studentsGraduateWithinProgramLength,
    studentsIntendedInFieldEmployment,
    employedInField90d,
    employedInField180d,
    notSeekingInField90d,
    notSeekingInField180d,
    stillSeekingInField90d,
    stillSeekingInField180d,
    medianAnnualBaseSalary90d,
    medianAnnualBaseSalary180d,
    under70K90d,
    under70K180d,
    _70K_80K90d,
    _70K_80K180d,
    _80K_90K90d,
    _80K_90K180d,
    over90K90d,
    over90K180d]

    mydf.loc[len(mydf)] = myrow



# Get the list of all files and directories
root_path = '/Users/robregon_ifds/Downloads'  ## YOU CAN CHANGE YOUR PATH HERE, WHERE YOU (O  N  L  Y)  HAVE YOUR PDFs from cirr.org/data
path = root_path + "/bootcamps/"  #CREATE THIS PATH
file_list = os.listdir(path)

cols = ['school_name',
        'campus_location',
        'program_name',
        'reporting_start', 
        'reporting_end', 
        'course_length',
        'graduates_included',
        'graduated_ontime',
        'intend_infield_employment',
        'employed_infield_90days',
        'employed_infield_180days',
        'not_seeking_infield_90days',
        'not_seeking_infield_180days',
        'still_seeking_infield_90days',
        'still_seeking_infield_180days',
        'median_annual_salary90days',
        'median_annual_salary180days',
        'under70K_90days',
        'under70K_180days',
        'salary_70K_80K_90days',
        'salary_70K_80K_180days',
        'salary_80K_90K_90days',
        'salary_80K_90K_180days',
        'over90K_90days',
        'over90K_180days']

#create a dataframe which will contain information from different bootcamps
mydf = pd.DataFrame(columns=cols)

#for loop for going through the files in the folder
for f in file_list:
    if '.pdf' in f:
        f2 = path + f
        reader = PdfReader(f2)

        # getting number of pages in pdf file
        pages = len(reader.pages)

        #rea the last page of the file, which contains the data to be extracted
        page = reader.pages[pages - 1]
        text = page.extract_text()

        parse_text(f2, text)


#save the dataframe to a resulting csv file
mydf.to_csv(path + '/bootcamps.csv')
```
