```python
import tabula
import pandas as pd
import numpy as np
import os
import re
  
folder_path = 'pdfs/data2'

# Create an empty list to hold the file names or paths
pdfs = []

# Loop through each file in the folder
for file_name in os.listdir(folder_path):
    # Check if the file is a PDF file
    if file_name.endswith('.pdf'):
        # Add the file name or path to the list
        file_path = os.path.join(folder_path, file_name)
        pdfs.append(file_path)  # Change to `file_name` to get only the file name instead of the full path

```

```python
dfs = []
for f in pdfs:
    # Use Tabula to extract pages with tables
    df = tabula.read_pdf(f, pages='all')[0]
    dfs.append(df)
```

```python
#check columns' names for each dataframe in the list of dataframes
for df in dfs:
    print(df.columns)
```

We are only interested in data frames with columns 'Report Information', 'Unnamed: 0', 'Unnamed: 1' or

```python
#pick the right dataframes
#only the dataframes that have 'Report Information' column
col_name = 'Report Information'
new_df = []
for df in dfs:
    if col_name not in df.columns:
        continue
    new_df.append(df)
```

```python
for df in new_df:
    if 'Unnamed: 1' in df.columns:
        df['Unnamed: 0'] = df['Unnamed: 0'].fillna('') + ' ' + df['Unnamed: 1'].fillna('')
        df.drop('Unnamed: 1', inplace=True, axis=1)
```

```python
for i in range(0, len(new_df)):
    if 'Unnamed: 2' in new_df[i].columns:
        new_df[i].drop('Unnamed: 2', axis=1, inplace=True)
```

```python
nums_toshift = [2,19,38,44]
for n in nums_toshift:
    new_df[n].loc[0, :] = new_df[n].loc[0, :].fillna('') +' '+ new_df[n].loc[1, :].fillna('')
    new_df[n] = new_df[n].drop(1)
```

```python
nums_toshift = [21,46,47,49]
for n in nums_toshift:
    if pd.isna(new_df[n].loc[2, 'Report Information']):
        new_df[n].loc[2,'Report Information'] = new_df[n].loc[3, 'Report Information']
        new_df[n].loc[2, 'Unnamed: 0'] = str(new_df[n].loc[2, 'Unnamed: 0']) + '' + str(new_df[n].loc[3, 'Unnamed: 0'])
    else:
        new_df[n].loc[2,'Report Information'] += new_df[n].loc[3, 'Report Information']
        new_df[n].loc[2, 'Unnamed: 0'] = str(new_df[n].loc[2, 'Unnamed: 0']) + ' ' + str(new_df[n].loc[3, 'Unnamed: 0'])
    new_df[n] = new_df[n].drop(3)
```

```python
#filter out all the rows containing 'days' substring in 'Report Information' column
# we are only interested in graduation rate
string_filter = 'days'
col_name = 'Report Information'
for i in range(0,len(new_df)):
    #replace NaN with empty string in order to use str.contains
    new_df[i] = new_df[i][new_df[i][col_name].str.endswith(string_filter) == False]
```

```python
substring = 'obtainers'
col_name = 'Report Information'
df2 = []
#create a regular expression to igonre case for the string I'm looking for
regex = re.compile(substring,re.IGNORECASE)
for i in range(0,len(new_df)):
    mask = new_df[i][col_name].apply(lambda x: bool(regex.search(str(x))))  # Create a boolean mask to filter rows containing the substring
    ind = new_df[i].loc[mask].index[0]  # Get the index of the first row containing the substring 
    new_df[i] = new_df[i].loc[:ind-1]
    
```

```python
#remove all rows that start with -, * or combination of digit and character
col_name = 'Report Information'
regex = re.compile(r'^([0-9][a-zA-Z]|[\\*\\-])')
for i in range(0,len(new_df)):
    mask = new_df[i][col_name].str.extract(regex).notnull().any(axis=1) # create a boolean mask
    new_df[i] = new_df[i][~mask]
    
```

```python
#transpose each dataframe, if the collumns appear null, change to 'col'
#select first row for columns names
new2 = []
for df in new_df:
    df = df.transpose()
    cols = df.iloc[0]
    cols.fillna('col', inplace=True)
    df.columns = cols
    df = df[1:]
    new2.append(df)
```

```python
for df in new2:
    s = df.columns.to_series().groupby(df.columns)
    df.columns = np.where(s.transform('size')>1, 
                      df.columns + s.cumcount().add(1).astype(str), 
                      df.columns)
```
