```python
import pandas as pd
import seaborn as sns

df  = pd.read_csv("./../data/combined_skills.csv")
```

```python
#remove % and $, so data can be imported into database
df = df.applymap(lambda x: x.replace('%', '') if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace('$', '') if isinstance(x, str) else x)
```

```python
#check if % and $ got removed
df
```

```python
# we have 'Codeup', 'Codeup, LLC' values in 'school_name' column
set(df['school_name'])
```

```python
#rename value
df['school_name'] = df['school_name'].replace(['Codeup, LLC'],['Codeup'])
```

```python
#check the program names 
set(df['program_name'])
```

```python
#rename programs' names in order to name consistency
rename_values = {'Front-End Web Developernan': 'Front-End Web Developer BootCamp', 'Front-End Web DeveloperBoot Camp': 'Front-End Web Developer BootCamp', 'Software Developer Bootcamp':'Software Developer BootCamp','Full-Stack Web Development -Java':'Full-Stack Web Development-Java','Data Science Boot Camp':'Data Science BootCamp','Data Science': 'Data Science BootCamp', 'Python Developer Boot Camp':'Python Developer BootCamp'}
df['program_name'] = df['program_name'].replace(rename_values)
```

```python
#rename all values containg 'remote' string into 'Remote'
rename_values = {'Central Remote Immersive': 'Remote','East Coast Remote':'Remote', 'National Live Remote':'Remote', 'Part Time Remote Immersive':'Remote','West Coast Remote':'Remote', 'Salt Lake City, Utah':'Salt Lake City','Tokyo, Japan':'Tokyo'}
df['campus_location'] = df['campus_location'].replace(rename_values)
```

```python
set(df['campus_location'])
```

```python
#create dictionary of cities and corresponding states
city_state= {'Boston': 'Massachusetts',
                   'Chicago': 'Illinois',
                   'Cincinnati': 'Ohio',
                   'Cleveland': 'Ohio',
                   'Columbus': 'Ohio',
                   'Dallas': 'Texas',
                   'Los Angeles': 'California',
                   'New York': 'New York',
                   'Online':'Online',
                   'Philadelphia': 'Pennsylvania',
                   'Pittsburgh': 'Pennsylvania',
                   'Salt Lake City': 'Utah',
                   'San Antonio': 'Texas',
                   'San Antonio - Vogue': 'Texas',
                   'Remote': 'Online',
                   'Tokyo': 'Japan',
                   'Toronto': 'Canada',
                   'Wilmington': 'Delaware'}
```

```python
# add new column state to the dataframe
df["state"] = df["campus_location"].apply(lambda x: city_state.get(x, None))
```

```python
set(df["state"])
```

```python
#temporarily replace missing values with -1 
df.fillna(-1, inplace=True)
```

```python
df.to_csv('E:/new/bootcamps_edu/data/combined_skills.csv',index=False)
```
